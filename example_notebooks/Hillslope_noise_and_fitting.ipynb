# Noise and fitting of hilltops

These are examples of how the hillslope toy fits profiles and add noise

```python
import hillslopetoy as ht
import matplotlib.pyplot as plt
```

```python
x = ht.set_profile_locations_half_length(half_length = 6,spacing = 1)
z = ht.ss_nonlinear_elevation(x,C_0=0.0004)
z = ht.set_channel_to_zero(z)

z_fit = ht.get_fitted_profile(x,z)
```

```python
plt.clf()
f, (ax1) = plt.subplots(1, 1)
ax1.plot(x, z, label="steady nonlinear",alpha = 0.8,color="r")
ax1.plot(x, z_fit, label="polynomial fit",alpha = 0.8,color="g")

# These are just the labels for the figure
ax1.set_xlabel("Distance from divide (m)",fontsize=24)
ax1.set_ylabel("Elevation (m)",fontsize=20)

for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)    

plt.legend(fontsize = 14)
plt.tight_layout()
```

## We can also add noise

```python
mean_uncert = 0.2
z_noise = ht.add_noise(z,mean_uncert)
z_noisy_fit = ht.get_fitted_profile(x,z_noise)

plt.clf()
f, (ax1) = plt.subplots(1, 1)
ax1.plot(x, z, label="steady nonlinear",alpha = 0.8,color="r")
ax1.plot(x, z_fit, label="polynomial fit",alpha = 0.8,color="g")
ax1.plot(x, z_noise, label="noisy version",alpha = 0.8,color="b")
ax1.plot(x, z_noisy_fit, label="noisy polynomial fit",alpha = 0.8,color="k")

# These are just the labels for the figure
ax1.set_xlabel("Distance from divide (m)",fontsize=24)
ax1.set_ylabel("Elevation (m)",fontsize=20)

for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)    

plt.legend(fontsize = 14)
plt.tight_layout()
```

We can also displace the divide to replicate when topographic data does not correspond exactly with the location of the divide.

```python
new_x = ht.displace_profile_locations_constant(x, displacement_distance = 0.5)
displace_z = ht.ss_nonlinear_elevation(new_x,C_0=0.0004)
displace_z = ht.set_channel_to_zero(displace_z)

mean_uncert = 0.2
noise_displace_z =  ht.add_noise(displace_z,mean_uncert)
z_noisy_displaced_fit = ht.get_fitted_profile(new_x,noise_displace_z)

plt.clf()
f, (ax1) = plt.subplots(1, 1)
ax1.plot(new_x, displace_z, label="displaced steady nonlinear",alpha = 0.8,color="r")
ax1.plot(new_x, noise_displace_z, label="noisy displaced steady",alpha = 0.8,color="g")
ax1.plot(new_x, z_noisy_displaced_fit, label="noisy displaced fit",alpha = 0.8,color="b")

# These are just the labels for the figure
ax1.set_xlabel("Distance from divide (m)",fontsize=24)
ax1.set_ylabel("Elevation (m)",fontsize=20)

for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)    

plt.legend(fontsize = 14)
plt.tight_layout()
```
