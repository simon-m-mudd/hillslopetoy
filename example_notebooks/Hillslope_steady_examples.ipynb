# Examples of hillslope plotting using the hillslope toy, steady state

For all of these examples, we need to first import the hillslope toy package, along with matplotlib for plotting

```python
import hillslopetoy as ht
import matplotlib.pyplot as plt
```

## Steady state hillslopes

Here are some examples of steady state hillslopes. First we plot linear hillslopes.
In these plots `C_0` is the uplift rate (using the symobols from early Roering et al. papers)

The default values are:
-    `C_0` = 0.001   uplift in m/yr
-   `D` = 0.001 the transport coefficient in m^2/yr
-   `rho_ratio` = 2 the ratio of soil and rock density (dimensionless)
-   `c` = 20 a constant that sets the elevation of the divide. We use another function to set the edge to zero after the profile is calculated. 

```python
x = ht.set_profile_locations_half_length(half_length = 30,spacing = 1)

z1 = ht.ss_linear_elevation(x,C_0=0.0001)
z1 = ht.set_channel_to_zero(z1)

z2 = ht.ss_linear_elevation(x,C_0=0.0002)
z2 = ht.set_channel_to_zero(z2)

z3 = ht.ss_linear_elevation(x,C_0=0.0004)
z3 = ht.set_channel_to_zero(z3)
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]

# Create two subplots and unpack the output array immediately
plt.clf()
f, (ax1) = plt.subplots(1, 1)
ax1.fill_between(x, 0,z1, label="U = 0.1 mm/yr",alpha = 0.8,facecolor="r")
ax1.fill_between(x, z1,z2, label="U = 0.2 mm/yr",alpha = 0.5,facecolor="r")
ax1.fill_between(x, z2,z3, label="U = 0.4 mm/yr",alpha = 0.2,facecolor="r")

# These are just the labels for the figure
ax1.set_xlabel("Distance from divide (m)",fontsize=24)
ax1.set_ylabel("Elevation (m)",fontsize=20)

for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)    

plt.legend(fontsize = 14)
plt.tight_layout()
```

Now lets plot some nonlinear hilltops.

The default values are:
-    `C_0` = 0.001   uplift in m/yr
-   `D` = 0.001 the transport coefficient in m^2/yr
-   `S_c` = 1 The critical slope value
-   `rho_ratio` = 2 the ratio of soil and rock density (dimensionless)
-   `c` = 20 a constant that sets the elevation of the divide. We use another function to set the edge to zero after the profile is calculated. 

```python
z1 = ht.ss_nonlinear_elevation(x,C_0=0.0001)
z1 = ht.set_channel_to_zero(z1)

z2 = ht.ss_nonlinear_elevation(x,C_0=0.0002)
z2 = ht.set_channel_to_zero(z2)

z3 = ht.ss_nonlinear_elevation(x,C_0=0.0004)
z3 = ht.set_channel_to_zero(z3)

# Create two subplots and unpack the output array immediately
plt.clf()
f, (ax1) = plt.subplots(1, 1)
ax1.fill_between(x, 0,z1, label="U = 0.1 mm/yr",alpha = 0.8,facecolor="r")
ax1.fill_between(x, z1,z2, label="U = 0.2 mm/yr",alpha = 0.5,facecolor="r")
ax1.fill_between(x, z2,z3, label="U = 0.4 mm/yr",alpha = 0.2,facecolor="r")

# These are just the labels for the figure
ax1.set_xlabel("Distance from divide (m)",fontsize=24)
ax1.set_ylabel("Elevation (m)",fontsize=20)

for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)    

plt.legend(fontsize = 14)
plt.tight_layout()
```

## Compare linear and nonlinear slopes

```python
z1 = ht.ss_linear_elevation(x,C_0=0.0004)
z1 = ht.set_channel_to_zero(z1)

z2 = ht.ss_nonlinear_elevation(x,C_0=0.0004)
z2 = ht.set_channel_to_zero(z2)

# Create two subplots and unpack the output array immediately
plt.clf()
f, (ax1) = plt.subplots(1, 1)
ax1.fill_between(x, 0,z1, label="linear",alpha = 0.8,facecolor="r")
ax1.fill_between(x, z1,z2, label="nonlinear",alpha = 0.5,facecolor="g")

# These are just the labels for the figure
ax1.set_xlabel("Distance from divide (m)",fontsize=24)
ax1.set_ylabel("Elevation (m)",fontsize=20)

for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)    

plt.legend(fontsize = 14)
plt.tight_layout()
```
